# DO KONKRETÓW - BŁĄDZENIE LOSOWE

1. cząstka o położeniu x (zaczynamy w 0)
2. ruszamy ją o losową z (-1 do 1)
3. zbieramy położenia do jakiejś tablicy
4. puszczamy rzędu 100000
5. plotujemy


````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def random(a, b):
    return a + (b - a) * np.random.random()

print(random(50,100))
    
x = 0

N = 100000
x_history = np.zeros(N)

for i in range(1,N):
    x = x + random(-1, 1)
    x_history[i] = x

plt.figure(figsize=(12,8))
plt.plot(x_history)
plt.xlabel("czas")
plt.ylabel("polozenie")
plt.title("random walk 1D")
plt.grid()
# plt.savefig("random_walk_1d.png") #zapis do pliku
````

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

NP = 50

def random(a, b):
    return a + (b - a) * np.random.random(NP)
x = np.zeros(NP)

N = 1000
x_history = np.zeros((N,NP))

for i in range(1,N):
    x = x + random(-1, 1)
    x_history[i] = x

plt.figure(figsize=(12,8))
plt.plot(x_history, "m-", alpha=0.3, linewidth=2)
plt.xlabel("czas")
plt.ylabel("polozenie")
plt.title("random walk 1D, {} czastek".format(NP))
plt.grid()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

NP = 3

def random(a, b):
    return a + (b - a) * np.random.random(NP)
x = np.zeros(NP)

N = 1000
x_history = np.zeros((N,NP))

for i in range(1,N):
    x = x + random(-1, 1)
    x_history[i] = x


fig = plt.figure(figsize=(9,6))
ax = fig.gca(projection='3d')
ax.plot(x_history[:,0], x_history[:,1], x_history[:,2], "g-", alpha=1, linewidth=1)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("random walk 3D, 1 czastka")
ax.grid()

plt.show()
#%matplotlib inline sprawia ze plot jest jako obrazek
#i nie mozna nim obracać
````

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import curve_fit

NP = 50

def random(a, b):
    return a + (b - a) * np.random.random(NP)
x = np.zeros(NP)

N = 1000
x_history = np.zeros((N,NP))

for i in range(1,N):
    x = x + random(-1, 1)
    x_history[i] = x

distances = np.sqrt(np.sum(x_history**2, axis=1))

def pierwiastek(x, A):
    return A*np.sqrt(x)

T =np.arange(N)

A, niepewnosc = curve_fit(pierwiastek, T, distances)
plt.plot(T, distances)
plt.plot(T, pierwiastek(T, A))
plt.show()
````

# Plotowanie Gaussianów do histogramu gaussowskich zmiennych losowych

````python
zmienne_losowe = np.random.normal(size=10000) #tworzę dane

y, x = np.histogram(zmienne_losowe, bins=50, density=True) #50 binów, znormalizować do 1
#uwaga kolejność! patrz dokumentacja np.histogram.
#UWAGA POPRAWKA: "DENSITY" ZAMIAST "NORMED", patrz dokumentacja

print(x.shape, y.shape)
y = y + np.random.random(y.shape)/100
x_left = x[:-1]   #x to tablica wszystkich kolejnych krawędzi, pomijając ostatni dostaję lewe krawędzie
bin_size = x[1] - x[0]
plt.bar(x_left,y, width = bin_size)       #bar bierze za x lewe krawędzie słupków, więc to jest ok

def Gauss(x, A, mean, std, srednia):
    return A * np.exp(-(x-mean)**2/2/std**2) + srednia

x_middle = (x_left + x[1:])*0.5      #liczę środki binów, bo do fitu to "bezpieczniej" 
parametry, kowariancje = curve_fit(Gauss, x_middle, y)
A, mean, std, srednia = parametry
odchylenia_standardowe = np.sqrt(np.diag(kowariancje))
dA = odchylenia_standardowe[0]


x_fit = np.linspace(x.min(), x.max(), 100)
y_fit = Gauss(x_fit, A, mean, std, srednia)
plt.plot(x_fit, y_fit, "m-", linewidth = 5)
plt.title("Amplituda: {:.3f} +- {:.3f}".format(A, dA))
plt.show()
````
