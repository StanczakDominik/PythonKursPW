# Random walk - N wymiarów, N cząstek

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#przypadek jednowymiarowy

x = 0
NT = 1000
x_history = np.zeros(NT)

for timestep in range(NT):
    x_history[timestep] = x         #zapisujemy dane "historyczne", żeby mieć co plotować
    x = x + np.random.random()*2 - 1   #krok o jednorodną losową z zakresu (-1, 1)

plt.plot(x_history)
plt.title("Random walk 1D")
plt.xlabel("Czas t")
plt.ylabel("x")
plt.show()
plt.hist(x_history, 50, normed=True)
plt.xlabel("x")
plt.ylabel("Rozklad prawdopodobienstwa x")
plt.show()
```

```python
#NWYMIAROWY RANDOM WALK! dla N = 2 i 1 cząstki

N = 2
x = np.zeros(N)
NT = 1000
x_history = np.zeros((NT, N))

for timestep in range(NT):
    x_history[timestep] = x
    x = x + np.random.random(N)*2 - 1

iks = x_history[:,0]
igrek = x_history[:,1]

plt.title("Random walk 2D - trajektoria pojedynczej czastki")
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.plot(iks, igrek)
plt.show()

plt.hist(iks, 50, normed=True, color = 'b', alpha=0.5, label="x")
plt.hist(igrek, 50, normed=True, color = 'r', alpha=0.5, label="y")
plt.legend()
plt.grid()
plt.xlabel("polozenie")
plt.ylabel("rozklad prawdopodobienstwa polozenia")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
#NWYMIAROWY RANDOM WALK N = 3

N = 3
x = np.zeros(N)
NT = 100000 #bo mozemy!
x_history = np.zeros((NT, N))

for timestep in range(NT):
    x_history[timestep] = x
    x = x + np.random.random(N)*2 - 1

    
fig = plt.figure()
axes = fig.gca(projection='3d')

iks = x_history[:,0]
igrek = x_history[:,1]
zet = x_history[:,2]

axes.plot(iks, igrek, zet, alpha=0.6)
plt.show()
```

```python
#NWYMIAROWY RANDOM WALK NP CZĄSTEK
#CZYLI JAK PRZESADZIĆ Z PRZYKŁADEM ŻEBY NIKT NIE OGARNĄŁ

N = 2
NP = 100
x = np.zeros(N*NP)
NT = 1000
x_history = np.zeros((NT, N*NP))

for timestep in range(NT):
    x_history[timestep] = x
    x = x + np.random.random(N*NP)*2 - 1

for czastka in range(NP):
    iks = x_history[:,czastka*N+0]
    igrek = x_history[:,czastka*N+1]
    plt.plot(iks, igrek, alpha=0.1)
# plt.savefig("wykres.png")
plt.show()
```
